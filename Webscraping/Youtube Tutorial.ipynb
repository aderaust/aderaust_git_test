Beautiful soup is used to parse text

urlib is used to accessing websites

```python
from bs4 import BeautifulSoup as soup

from urllib.request import urlopen as uReq
```

```python
# grabing url for parse to work with

my_url = 'https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=graphics%20card'
```

```python
# opening up connection then downloading the data from this page
uClient = uReq(my_url)
```

```python
# reading the raw html

page_html = uClient.read()

# closing the client or connection
uClient.close()
```

```python
page_soup = soup(page_html, 'html.parser')

page_soup
```

```python
page_soup.body.a
```

```python
# parsing all div tags that contain an item-container

containers = page_soup.findAll('div', {'class':'item-container'})
len(containers)
```

```python
# page_soup.find()
price_info = containers[0].find('div', {'class':'item-action'})
```
